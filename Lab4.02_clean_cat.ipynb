```python
def clean_policy(x):
    if 'Personal' in x:
        return 'Personal'
    elif 'Corporate' in x:
        return 'Corporate'
    else:
        return 'Special'

categorical_df['policy_large'] = categorical_df['Policy'].apply(clean_policy)
categorical_df.head(20)
```
